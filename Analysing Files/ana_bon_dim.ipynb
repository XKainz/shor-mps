```python
import pickle
import matplotlib.pyplot as plt
import numpy as np
```

```python
bond_dim_array = pickle.load(open("./pickles/bond_dim_array_299_300.pkl","rb"))
bond_dim_array = np.array(bond_dim_array)
plt.figure(figsize=(15,7))
plt.plot(bond_dim_array[:,0],bond_dim_array[:,2],label="MPO")
plt.plot(bond_dim_array[:,0],bond_dim_array[:,4],label="MPS")
plt.xlabel("N")
plt.ylabel("Bond Dimension")
plt.legend()
plt.show()
```
