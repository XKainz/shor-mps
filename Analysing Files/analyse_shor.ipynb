```python
import pickle
import numpy as np
import matplotlib.pyplot as plt
```

```python
def plot_p_success_rates(to_pickle):
    p_suc_mpo = to_pickle.p_success_mpo
    p_suc_fat = to_pickle.p_success_fat
    plt.figure(figsize=(15,7))
    plt.plot(p_suc_mpo[:,0],p_suc_mpo[:,1],label="MPO")
    plt.plot(p_suc_fat[:,0],p_suc_fat[:,1],label="FAT")
    plt.legend()
    plt.xlabel("Max Bond Dimension")
    plt.ylabel("Successful factorization rate")
    plt.title("Successful factorization rate vs. Max Bond Dimension")
```

```python
with open("./pickles/to_pickle51_40.pkl","rb") as f:
        to_pickle = pickle.load(f)
to_pickle.MPO_mpo.plot_schmidt_values("Schmidt values for MPO")
#to_pickle.MPS_fat.plot_schmidt_values("Schmidt values for MPS fat")
to_pickle.MPS_mpo.plot_schmidt_values("Schmidt values for MPS mpo")
print(to_pickle.max_bond_dim_mps_mpo,"mps_mpo")
print(to_pickle.max_bond_dim_mps_fat,"mps_fat")
print(to_pickle.max_bond_dim_mpo_mpo,"mpo_mpo")

to_pickle.MPO_mpo.plot_entanglement("Entanglement of MPO")
to_pickle.MPO_mpo.plot_bond_dims("Bond dimensions of MPO")
#to_pickle.MPS_fat.plot_entanglement("Entanglement of MPS fat")
to_pickle.MPS_mpo.plot_entanglement("Entanglement of MPS mpo")
to_pickle.MPS_mpo.plot_bond_dims("Bond dimensions of MPS mpo")

plot_p_success_rates(to_pickle)
```
