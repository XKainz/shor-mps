```python
import numpy as np
import matplotlib.pyplot as plt
import pickle
import Gates as gts
import shor_helpers as sh
from MPO import *
from SuperMPS import *
```

```python
class ana_bond_dim_mpo_obj(object):
    def __init__(self,
                 schmidt_values,
                 N,
                 x):
        self.schmidt_values = schmidt_values
        self.N = N
        self.x = x
```

```python
xi_start = 2**13
pickle_objs =[]
start = 15
end = 1000
path_to_pickles = "/space/ge65kox/BA/pickles/"
path_to_pickles = path_to_pickles+"analyse_bond_dimensions_mpo_"+str(start)+"_"+str(end)+".pkl"
for i in range(start,end):
    if sh.N_valid(i):
        success = 0
        while success == 0:
            x = sh.get_x_for_N(i)
            try:
                mpo = gts.cx_pow_2k_mod_N_mpo_from_fatU(i,x,1,xi_start)
                schmidt_values = mpo[0].get_schmidt_values_all_sites()
                pickle_objs.append(ana_bond_dim_mpo_obj(schmidt_values,i,x))
                success = 1
            except Exception as e:
                print(e)
                success = 0
        pickle.dump(pickle_objs,open(path_to_pickles,"wb"))
```

```python
def get_max_bond_dimension(schmidt_values):
    l = [len(s) for s in schmidt_values]
    return np.max(l) 
```

```python
#unpickle
pickle_objs = pickle.load(open(path_to_pickles,"rb"))

site = [len(pickle_objs[i].schmidt_values) -1 for i in range(len(pickle_objs))]
max_bond_dim_mpo = np.array([get_max_bond_dimension(pickle_objs[i].schmidt_values) for i in range(len(pickle_objs))])
print(site)
```

```python
N = np.array([pickle_objs[i].N for i in range(len(pickle_objs))])
x = np.array([pickle_objs[i].x for i in range(len(pickle_objs))])

#calculate the order of x^r mod N
def calculate_order(x,N):
    r = 1
    while pow((int)(x),r,mod = (int)(N)) != 1:
        r += 1
    return r

order = [calculate_order(x[i],N[i]) for i in range(len(N))]
```

```python
#plot
plt.figure(figsize=(17,10))
plt.scatter(order,max_bond_dim_mpo)
plt.xlabel("order of x^r mod N")
plt.ylabel("max bond dimension")
plt.show()


plt.figure(figsize=(17,10))
plt.scatter(N,max_bond_dim_mpo)
plt.xlabel("N")
plt.ylabel("max bond dimension")
plt.show()
```

```python
#Bond dimensions of random Unitaries
def get_random_unitary(d_sites):
    d = 2**d_sites
    A = np.random.rand(d,d)+1j*np.random.rand(d,d)
    Q,R = np.linalg.qr(A)
    return Q

random_unitary = get_random_unitary(3)
print()
mpo = MPO.create_MPO_from_tensor(random_unitary.reshape((2,)*2*3),xi=2**13,cutoff=1e-10)
sv = mpo.get_schmidt_values_all_sites()
mpo.plot_schmidt_values()



schmidt_values = []
size = []
for i in range(11):
    for j in range(4):
        mpo = MPO.create_MPO_from_tensor(get_random_unitary(i).reshape((2,)*2*i),xi=2**13,cutoff=1e-10)
        print(mpo.L)
        #mpo.plot_schmidt_values()
        schmidt_values.append(mpo.get_schmidt_values_all_sites())
        size.append(2**i-1)


max_bond_dim_random_unitary = np.array([get_max_bond_dimension(schmidt_values[i]) for i in range(len(schmidt_values))])
```

```python
plt.figure(figsize=(17,10))
plt.scatter(N,max_bond_dim_mpo,label ="x mod N mpo")
plt.scatter(size,max_bond_dim_random_unitary,label="random unitary")
plt.xlabel("N")
plt.ylabel("max bond dimension")
plt.legend()
plt.show()
```
