```python
import pickle
import numpy as np
from analysing_xi import *
import matplotlib.pyplot as plt
import tim 

path_to_pickles = "/space/ge65kox/BA/pickles/"
path_to_pickles = path_to_pickles+"ana_ent_objs_"
start = 20
end = 10000
xi_mpo = 64
xi_mps = 8
path_to_pickles = path_to_pickles+str(start)+"_"+str(end)+"_xi_mpo_"+str(xi_mpo)+"_xi_mps_"+str(xi_mps)+"_2.pkl"

```

```python
#get the data
objs = pickle.load(open(path_to_pickles,"rb"))
N = [obj.N for obj in objs]
x = [obj.x for obj in objs]
p_success_mpo = [obj.p_success_mpo for obj in objs]
time_to_create_mpos = [obj.tim_create_mpos[-1] - obj.tim_create_mpos[0] for obj in objs]
xi_max_in_mpos = [np.max(obj.xi_max_in_mpos) for obj in objs]
xi_max_in_mps = [obj.xi_max_in_mps for obj in objs]
print("N: ",N)
print("x: ",x)
```

```python
#plot times
plt.plot(N,time_to_create_mpos)
plt.xlabel("N")
plt.ylabel("time to create mpos [s]")
plt.show()
```

```python
plt.plot(N,p_success_mpo)
plt.xlabel("N")
plt.ylabel("p_success_mpo")
plt.show()
```

```python
#plot xi_max
plt.plot(N,xi_max_in_mpos,label="xi_max_in_mpos")
plt.plot(N,xi_max_in_mps,label="xi_max_in_mps")
plt.xlabel("N")
plt.ylabel("xi_max")
plt.legend()
plt.show()
```
